```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, to_date, hour, dayofweek, expr, unix_timestamp
from pyspark.sql.window import Window
from pyspark.sql import functions as F

# Create a Spark session
spark = SparkSession.builder.appName("CitiBikePrediction").getOrCreate()

# Set the path to your downloaded CitiBike data
citi_bike_data_path = "file:/tmp/*.csv"

# Read all CSV files into a Spark DataFrame and aggregate into daily records
citi_bike_df = spark.read.format("csv").option("header", True).load(citi_bike_data_path)

# Create a new column for datetime only
citi_bike_df = citi_bike_df.withColumn("datetime", col("started_at"))

# Convert datetime column to a standard date format
citi_bike_df = citi_bike_df.withColumn("datetime", to_date("datetime", "yyyy-MM-dd HH:mm:ss"))

# Target column
daily_trip_count_df = citi_bike_df.groupBy("datetime").agg(F.count("ride_id").alias("daily_trip_count"))

# Features
daily_trip_count_df = daily_trip_count_df.withColumn("day_of_week", dayofweek("datetime"))

# Intermediate DataFrames
i_daily_proc_df = citi_bike_df.withColumn("trip_duration", (unix_timestamp("ended_at", "yyyy-MM-dd HH:mm:ss") - unix_timestamp("started_at", "yyyy-MM-dd HH:mm:ss")))
i_daily_proc_df = i_daily_proc_df.withColumn(
    "trip_distance",
    F.sqrt((col("end_lng") - col("start_lng"))**2 + (col("end_lat") - col("start_lat"))**2)
)

i_daily_avg_df = i_daily_proc_df.groupBy("datetime").agg(
    F.mean("trip_duration").alias("avg_trip_duration"),
    F.mean("trip_distance").alias("avg_trip_distance"),
    (F.sum(F.when(col("rideable_type") == "electric_bike", 1).otherwise(0)) / F.count("*")).alias("avg_electric"),
    (F.sum(F.when(col("member_casual") == "member", 1).otherwise(0)) / F.count("*")).alias("avg_member")
)

# Cumulative Average Features
# Define a Window specification ordered by datetime
cumulative_window = Window.orderBy("datetime").rowsBetween(Window.unboundedPreceding, Window.currentRow)
# Perform aggregation with cumulative window
cumulative_avg_features_df = i_daily_avg_df.withColumn(
    "cumulative_avg_trip_duration",
    F.avg("avg_trip_duration").over(cumulative_window)
).withColumn(
    "cumulative_avg_trip_distance",
    F.avg("avg_trip_distance").over(cumulative_window)
).withColumn(
    "cumulative_avg_electric",
    F.avg("avg_electric").over(cumulative_window)
).withColumn(
    "cumulative_avg_member",
    F.avg("avg_member").over(cumulative_window)
)
# Drop duplicate columns from i_daily_avg_df
cumulative_avg_features_df = cumulative_avg_features_df.drop(
    "avg_trip_duration", "avg_trip_distance",
    "avg_electric", "avg_member"
)

# Show the result
display(cumulative_avg_features_df)

# 7-day Moving Average Features
# Define a Window specification ordered by datetime
moving_window = Window.orderBy("datetime").rowsBetween(-6, 0)  # 7-day moving average
# Perform aggregation with moving window
moving_avg_features_df = i_daily_avg_df.withColumn(
    "moving_avg_trip_duration",
    F.avg("avg_trip_duration").over(moving_window)
).withColumn(
    "moving_avg_trip_distance",
    F.avg("avg_trip_distance").over(moving_window)
)
# Drop duplicate columns from i_daily_avg_df
moving_avg_features_df = moving_avg_features_df.drop(
    "avg_trip_duration", "avg_trip_distance",
    "avg_electric", "avg_member"
)

# Join daily_trip_count_df with cumulative_avg_features_df on "datetime"
final_citibike_df = daily_trip_count_df.join(cumulative_avg_features_df, on="datetime", how="left")

# Join the result with moving_average_features_df on "datetime"
final_citibike_df = final_citibike_df.join(moving_avg_features_df, on="datetime", how="left")

# Display the final DataFrame
display(final_citibike_df)
```

## 2. Build ML Pipeline using Training Data

```python
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler, StandardScaler
from pyspark.ml.regression import RandomForestRegressor

# Define categorical features
categorical_features = ["day_of_week"]

# Define numerical features
numerical_features = ["cumulative_avg_trip_duration", "cumulative_avg_trip_distance", "cumulative_avg_electric","cumulative_avg_member", "moving_avg_trip_duration", "moving_avg_trip_distance", "precip", "precipcover", "feelslike", "snowdepth", "windgust", "windspeed","cloudcover", "visibility", "solarradiation", "uvindex", "severerisk"]

# Step 1: Encode categorical features
indexers = [StringIndexer(inputCol=col, outputCol=col + "_index", handleInvalid="skip") for col in categorical_features]

# Step 2: Assemble features into a single vector
assembler = VectorAssembler(inputCols=[col + "_index" for col in categorical_features] + numerical_features, outputCol="features", handleInvalid="skip")

# Step 3: Scale down numerical features
scaler = StandardScaler(inputCol="features", outputCol="scaled_features", withStd=True, withMean=False)

# Step 4: Build the machine learning model (Random Forest Regressor in this case)
rf = RandomForestRegressor(featuresCol="scaled_features", labelCol="daily_trip_count", numTrees=100, seed=42)

# Create the pipeline
pipeline = Pipeline(stages=indexers + [assembler, scaler, rf])

# Train the pipeline using the training data
try:
    model = pipeline.fit(training_data)
except Exception as e:
    print(f"Error: {e}")
```

## 3. Evaluate ML Pipeline using Testing Data

```python
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml import PipelineModel

# Make predictions on the testing data
predictions = model.transform(testing_data)

# Evaluate the performance using RegressionEvaluator
evaluator = RegressionEvaluator(labelCol="daily_trip_count", predictionCol="prediction", metricName="mae")
mae = evaluator.evaluate(predictions)

evaluator_r2 = RegressionEvaluator(labelCol="daily_trip_count", predictionCol="prediction", metricName="r2")
r2 = evaluator_r2.evaluate(predictions)

# Display the evaluation metrics
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared value: {r2}")
```
